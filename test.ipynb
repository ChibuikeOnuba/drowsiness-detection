```python
data['rounded_ear'] = data['ear_main'].apply(lambda x: round(x,1))
data['rounded_mar'] = data['mar_main'].apply(lambda x: round(x,1))
data
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
data['time'] = pd.to_datetime(data['time'])
seconds = data['time'].dt.second

# Plot ear_main against time in seconds
plt.plot(seconds, data['ear_main'], linestyle='--', linewidth=0.5, color='red', label='EAR')
plt.fill_between(seconds, data['ear_main'], linestyle='--', linewidth=0.5, color='red', alpha=0.3)

plt.xlim(0, 30)
# You might want to adjust the labels and title accordingly
plt.xlabel('Time (Seconds)')
plt.ylabel('EAR')
plt.title('EAR over Time (Seconds)')
plt.legend()
plt.show()
```

```python
# Assuming 'data' is your DataFrame containing the relevant data
# Convert 'time' column to datetime datatype
# Assuming 'data' is your DataFrame containing the relevant data
# Convert 'time' column to datetime datatype
data['time'] = pd.to_datetime(data['time'])

# Set the size of the figure
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_ylim(-10,10)
# Plot ear_main against time with seconds
ax.plot(data['time'], data['ear_main'], linestyle='--', linewidth=0.5, color='red', label='EAR')

# Format x-axis to show minutes with seconds
ax.xaxis.set_major_formatter(mdates.DateFormatter('%M:%S'))

# Set x-axis locator to every 10 seconds
ax.xaxis.set_major_locator(mdates.SecondLocator(interval=10))

# You might want to adjust the labels and title accordingly
ax.set_xlabel('Time (Minutes:Seconds)')
ax.set_ylabel('EAR')
ax.set_title('EAR over Time')
ax.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

min_ear_time = data[data['ear_main'] == data['ear_main'].min()]['time']
hour_minute = min_ear_time.dt.strftime('%H:%M').values[0]

EAR_THRESHOLD = -0.4
list_lenght = []
for i in (data['ear_main']).tolist():
    if i < EAR_THRESHOLD:
        list_lenght.append(i)
    

# Add text beside the plot
text = f'''
EAR THRESHOLD: -0.4
MAR THRESHOLD:
Time you felt most drowsy: {hour_minute}
beside the plot to provide additional information.
You can customize the content and formatting as needed.
Number of times you felt drowsy: {len(list_lenght)}
    '''
plt.text(1.01, 0.02, text, transform=ax.transAxes, verticalalignment='bottom', fontsize=8)

plt.show()
```
